```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Z=1
a=1
x=np.arange(0,12)
y=(1/np.sqrt(2))*(Z/a**3/2)*(1-Z*x/2*a)*np.exp(-Z*x/2*a)
plt.plot(x,y,color="purple")
plt.title("RADICAL DEL HIDRÃ“GENO")
plt.grid()
plt.show()
```
