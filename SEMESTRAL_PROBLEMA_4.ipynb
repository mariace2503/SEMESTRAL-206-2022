```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url="https://raw.githubusercontent.com/mariace2503/QM206MARIACORDOBA/main/CHIRIQUI-DISTRITOS%20-%20Hoja%201.csv"
df=pd.read_csv(url)
Población=df["Población"]
Distrito=df["Distrito"]
plt.barh(Distrito,Población,color="blue")
plt.xlabel("Población")
plt.ylabel("Distrito")
plt.title("POBLACIÓN DE LOS DISTRITOS DE CHIRIQUÍ")
plt.show()
```
